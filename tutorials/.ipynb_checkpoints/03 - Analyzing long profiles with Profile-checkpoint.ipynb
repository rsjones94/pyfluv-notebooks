# Analyzing long profiles with the Profile class

## Initializing a Profile

The `Profile` class relatively few arguments compared to the `Profile` class:

`Profile(df, name=None, metric=False)`

Only `df` is required. `df` is either a dictionary or pandas dataframe with mandatory the columns/keys `'exes'`, `'whys'` and `'Thalweg'`. `'exes'` and `'whys'` are planform arguments; `'Thalweg'` is an elevation argument. Other columns/keys are optional but the following have a specific meaning to the `Profile` class: `'Water Surface'`, `'Bankfull'`, `'Top of Bank'`, `'Riffle'`, `'Run'`, `'Pool'`, `'Glide'`. You should *not* pass df with a column/key called `'Station'`; this will be calculated upon initialization. If a dictionary is passed it will be coerced to a pandas dataframe.

We can make a simple Profile:

```python
import pyfluv

x = [1,20,40,30,60,60]
y = [0,0,20,20,40,80]
z = [5,4,2,1,2,2]
ws = [6,5,3,2,3,3]

surveyDict = {'exes':x,
              'whys':y,
              'Thalweg':z,
              'Water Surface':ws}

myPro = pyfluv.Profile(surveyDict, name='myPro')
```

Like CrossSection objects, Profile objects have quick plots and planform plots.

```python
import matplotlib.pyplot as plt

plt.figure()
myPro.qplot()

plt.figure()
myPro.planplot()
```

## Working with Profiles

Like with cross sections, it's difficult to make interesting synthetic profile data so profiles from the West Piney monitoring project are included in pyfluv.

```python
survey = pyfluv.piney_survey()
pros = survey.get_profile_objects()
```

These profiles don't throw any errors upon initialization, but a warning will be raised if an isolated riffle, run, pool or glide call is made. An isolated substrate call is one where neither the preceding nor following substrate call is the same substrate type. Such a feature thus begins and ends at the same point and has a length of zero. These calls are removed from `df` upon initalization.

An exception will be raised if a single substrate call has more than two substrate types. Up to two are permitted, as two simultaneous calls indicate one feature ending where another begins.

```python
import matplotlib.pyplot as plt

plt.figure()
pros[4].qplot(ve=60)  # like CrossSections, a vertical exaggeration can be specified with ve

plt.figure()
pros[4].planplot()
```

Unlike `CrossSection` objects, the fundamental cache of data for `Profile` objects is stored in `filldf`, not `df`. `df` is merely a copy of what was passed to the `df` parameter on initialization.

```python
pros[4].filldf
```

A few things to notice. First, `StreamSurvey` has filled the `'desc'` column with the descriptions associated with each row in a list (since some rows may represent multiple shots, e.g., a substrate and subsequent water shot). Second, the `'Water Surface'`, `'Bankfull'` and `'Top of Bank'` columns have been interpolated.

```python
import pandas as pd

c1 = pros[3].filldf['Top of Bank']
c2 = pros[3].df['Top of Bank']

pd.DataFrame([c1,c2]).transpose()
```

Third, a few columns have been created using the data passed on initalization: `'Water Depth'`, `'Bankfull Height'`, `'Bankfull to Water'` and `'Top of Bank Height'`.

Because we have riffle and pool calls (though no runs or glides), we can visualize them in the qplot and planplot.

```python
plt.figure()
pros[4].qplot(showFeatures=True)

plt.figure()
pros[4].planplot(showFeatures=True)
```

These features are stored in a dictionary that is suitably named `features`.

```python
pros[4].features
```

The `Feature` class is a subclass of `Profile`, and can generally be interacted with in the same way that you would interact with a `Profile`, though it does not have a `filldf` attribute and has an additional `morphType` attribute. `morphType` is a string and can be `'Riffle'`, `'Run'`, `'Pool'`, `'Glide'` or `Unclassified`. Generally you should not modify a `Feature` object except via its parent `Profile` object's method `reclassify_feature()`

## Extracting statistics

`Profile` (and `Feature`) objects have a a few methods included with them that calculate statistics relevant to restoration geomorphology.

```python
import numpy as np

print(f'Length is {pros[4].length()}')
print(f'Sinuosity is {pros[4].sinuosity()}')
print(f'Mean water slope is {pros[4].mean_slope("Water Surface")}')
print(f'Mean bankfull slope is {pros[4].mean_slope("Bankfull")}')

bkfEq = pros[4].fit(col='Bankfull')
print(f'Bankfull equation is y = {bkfEq[0]}x + {bkfEq[1]}')

print(f'Pool spacing is {np.mean(pros[4].spacing(featureType="Pool"))}')
```

The `mean_slope()` method can be called on any column in `filldf` that has an entry in the first and last row. The `featureType` argument in `spacing()` can be any valid key in `features`. `spacing()` returns returns a list of the distance between each subsequent `Feature` in the in the list contained in `features(featureType)`. Thus the method will always return a list of length `n-1` where `n` is the length of the list that `featureType` points to.

There are also optional `spacingFrom` and `spacingTo` arguments. By default these are both `'deepest'`, but they can also be `'start'`, `'end'`, or `'middle'`. If either is `'deepest'`, then another parameter `'deepType'` is used. By default it is `'Water Depth'`, meaning the deepest point is considered the point where the water depth is maximized. It can also be `'Bankfull Height'` or `'Thalweg'`. When these values are passed the deepest point will be respectively where the bankfull height is maximized or the thalweg elevation is minimized.

`fit()` returns the coefficients of a linear regression for a column in `filldf`. By default, the regression is a 1st order (linear) polynomial. An optional `order` argument allows you to return a higher (or 0th) order polynomial. The coefficients are returned in decreasing order.

Because `Feature` is a subclass of `Profile`, you can call all of the above methods on `Feature` objects, though to use `spacing()` you would need to deliberately call the functions to make a `features` dictionary beforehand.

```python
feat = pros[4].features['Riffle'][1]

print(f'Length is {feat.length()}')
print(f'Sinuosity is {feat.sinuosity()}')
print(f'Mean water slope is {feat.mean_slope("Water Surface")}')
print(f'Mean bankfull slope is {feat.mean_slope("Bankfull")}')

bkfEq = feat.fit(col='Bankfull')
print(f'Bankfull equation is y = {bkfEq[0]}x + {bkfEq[1]}')

plt.figure()
feat.qplot()
```

You can also show the regression lines on a qplot by calling `trend()` method. It takes the same arguments as `fit()`.

```python
plt.figure()
pros[4].qplot()
pros[4].trend('Water Surface')
pros[4].trend('Bankfull',2)
pros[4].trend('Top of Bank', 0)
```

## Modifying and predicting features

Sometimes we need to change our calls on a feature. You can do this in two ways. The direct way is to modify the entries in `filldf`. You will need to call the `create_features()` method afterwards to make sure all of your data is in agreement. It is also a good idea to call `validate_substrate()` beforehand to make sure you haven't inadvertently created zero-length features or rows with three or more substrate calls. As a general rule, it is not recommended to directly modify values unless you are confident in what you are doing.

```python
pros[4].filldf.loc[27:40,'Pool'] = np.NaN
pros[4].filldf.loc[27:40,'Riffle'] = pros[4].filldf.loc[27:40,'Thalweg'] 

pros[4].validate_substrate()
pros[4].create_features()

plt.figure()
pros[4].qplot(showFeatures=True)
```

Once a feature begins, it is continued until the subsequent row value for that substrate type is a null value. Because of this it is important to be aware of the impact that modifying `filldf` might have in terms of joining two previously distinct features. An exception to this continuity rule occurs when another substrate type punctuates the continuity. For example, in the following situation a riffle made of two points would be identified, followed by a pool made of type points, followed by a riffle made of two points. If the pool column were entirely null values, there would be one riffle made of four points.

```python
ri = pd.Series(data=[90,89,88,87], name='Riffle')
po = pd.Series(data=[np.NaN,89,88,np.NaN], name='Pool')

pd.DataFrame([ri,po]).transpose()
```

You can also reclassify features in `features` directly with the `reclassify_feature()` method. The method accepts a `Feature` object for its `feature` argument and a valid substrate type for its `newMorph` argument. By default it updates the `features` dictionary and modifies the substrate columns in `filldf` for you. This is particularly useful for changing `'Unclassified'` features.

```python
targetFeat1 = pros[4].features['Unclassified'][0]
pros[4].reclassify_feature(targetFeat1,newMorph='Run')

plt.figure()
pros[4].qplot(showFeatures=True)
```

You can reclassify all `'Unclassified'` features at once with the `classify_by_adjacency()` method.

```python
plt.figure()
pros[5].qplot(showFeatures=True)

pros[5].classify_by_adjacency()
plt.figure()
pros[5].qplot(showFeatures=True)
```

What if you don't have any data about substrate morphology? `feature_classify_k_means()` can predict features using water depth and water slope data. We can compare the real field calls to what the method predicts.

```python
plt.figure()
pros[5].qplot(showFeatures=True)

pros[5].feature_classify_k_means()
plt.figure()
pros[5].qplot(showFeatures=True)
```

This result isn't too far off from what was observed in the field. This tool can be helpful when working with a large amount of survey data that is lacking field calls.
